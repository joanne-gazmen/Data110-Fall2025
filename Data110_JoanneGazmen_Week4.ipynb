## Task 1: Basic Bar Chart for Crime Rates in 2023

In this task, you will create a **bar chart** showing the offense rates per 100,000 people for each type of crime in 2023.

### Instructions:
- Create a bar chart for crime rates in 2023.
- Customize the bar colors and add gridlines.
- Label the axes and add a title.

````python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
print(plt.style.available)
plt.style.use("seaborn-v0_8-white")  # Sets the style for the plots
````

````python
x_crimes = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
y_rates = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]

# Bar chart
plt.figure(figsize=(10, 5))  # Set the figure size so all names on the x-axis fit
plt.bar(x_crimes, y_rates, color = '#68f28d') #plot graph with custom color

# Labels, title, grid
plt.title('Offenses per 100,000 People(2023)', fontweight='bold', fontsize=16)
plt.xlabel('Offense Type', fontsize=10, fontweight='bold')
plt.ylabel('Offenses per 100,000 People', fontsize=10, fontweight='bold')
plt.grid(axis='y', linestyle='--',color = 'gray') #add gridlines


# Show the plot
plt.show()
````

## Task 2: Change in Crime Rates (2022 vs. 2023) with Unique Colors

In this task, you will create a **horizontal bar chart** that visualizes the **percentage change** in offense rates from 2022 to 2023 for each crime type. Each bar should have a unique color for better distinction.

### Instructions:
- Create a horizontal bar chart showing the **change in offense rates**.
- Assign a unique color to each crime type.
- Label the axes and add a title.

````python
offenses = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
percentage_change = [-3.3       ,    -8.1   ,    -4.9  ,          +12.0       ,    -12.0,   -9.8,    -0.8  ]
colors = ['#5c53a5','#a059a0','#ce6693','#eb7f86','#f8a07e','#fac484','#f3e79b']

# bar chart
plt.figure(figsize=(12, 4))
plt.barh(offenses, percentage_change, color=colors, height=0.7)

# labels
plt.title('Change in Offenses per 100,000 People (2022-2023)', fontweight = 'bold', size = 20)
plt.xlabel('Percent Change (by 100,000 people)', fontweight='bold', size = 15)
plt.ylabel('Offense', fontweight='bold', size = 15)

# grid and horizontal line
plt.grid(axis='x', linestyle='solid')
plt.grid(axis='y', linestyle='solid')
plt.axvline(x=0, color='black') # plot a vertical line at 0 to indicate where the horizontal graph starts


#Show the plot
plt.show()
````
